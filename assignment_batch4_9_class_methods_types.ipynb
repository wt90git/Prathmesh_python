## Q1: Create a class called Student

Requirements:
- Class variable: `school_name = "ABC International School"`
- Constructor: takes `name` and `marks`
- Instance method: `display_details()`
- Class method: `change_school()`
- Static method: `is_pass(marks)` returns "Pass" if marks â‰¥ 35 else "Fail"


```python
class Student:
    school_name = "ABC International School"

    def __init__(self, name, marks): #constructor Initiated
        self.name = name
        self.marks = marks

    def display_details(self):
        print("Name:",self.name," Marks",self.marks,"School: ",Student.school_name)

    @classmethod
    def change_school(cls, new_school):
        cls.school_name = new_school
    
    @staticmethod
    def is_pass(marks):
        return "Pass" if marks >=35 else "Fail"
       
```

```python
s1 = Student("Rahul", 89)
```

```python
s1.display_details()
```

```python
Student.change_school("Modern International School")
```

```python
s1.display_details()
```

```python
s1.is_pass(s1.marks)
```

```python
Student.is_pass(90)
```

## Q2: Create a class Calculator

Requirements:
- Class variable: `count_operations = 0`
- Instance method: `add(a, b)`
- Class method: `increment_count()`
- Static method: `info()`


```python

class Calculator:
    count_operation = 0

    def addition(self, a, b):
        Calculator.increment_count()
        return(a + b)

    @classmethod
    def increment_count(cls):
        cls.count_operation += 1  

    @staticmethod
    def info():
        print("Total Operation:",Calculator.count_operation)   
```

```python
Calculator.count_operation
```

```python
op1 = Calculator()
```

```python
op2 = Calculator()
```

```python
op1.addition(4,6)
```

```python
op2.addition(2,2)
```

```python
Calculator.count_operation
```

```python
op1.info()
```

## Q3: Write a class BankAccount

Requirements:
- Class variable: `bank_name = "SBI"`
- Constructor: `name`, `balance`
- Instance method: `deposit(amount)`
- Class method: `change_bank(new_name)`
- Static method: `validate_amount(amount)`


```python
class BankAccount:
    bank_name = "SBI"

    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        if BankAccount.validate_amount(amount):
            self.balance+=amount
            print(f"Rs {amount} is deposited in your acc. Balance {self.balance}")
        else:
            print("Invalid Amount")
       
      
    
    @classmethod
    def change_name(cls, new_bank):
        cls.bank_name = new_bank
    
    @staticmethod
    def validate_amount(amount):
        return amount > 0
```
