**Q24.** Make a class `Date` with class method `from_string(cls, s)` to parse `"DD-MM-YYYY"` and return an object.

```python
# TODO: Write your solution for Q24 below
class Date:

    def __init__(self,day,month,year):
        self.day = day
        self.month = month
        self.year = year

    @classmethod
    def from_strings(cls,s):
        day, month, year = s.split("-")
        return cls(int(day), int(month), int(year))
    
    def show(self):
        print(f"{self.day}-{self.month}-{self.year}")

```

```python
d1 = Date.from_strings("10-06-25")
print(d1.day, d1.month, d1.year)
```

```python
d1.show()
```

## G. Decorators (27–30)


**Q27.** Write a simple decorator that prints `"Function started"` before running the function. Test it on a small function.

```python
# TODO: Write your solution for Q27 below
def start_decorator(func):
    def wrapper():
        print("Function started")
        return func()
    return wrapper
```

```python
@start_decorator
def greet():
    print("Hello!")
```

```python
greet()
```

**Q28.** Create a decorator that prints `"Function ended"` after running the function. Test it.

```python
# TODO: Write your solution for Q28 below
def end_decorator(func):
    def wrapper():
        result = func()
        print("Function Ended")
        return result
    return wrapper
```

```python
@end_decorator
def greet():
    print("Hello!")
```

```python
greet()
```

**Q29.** Make a decorator that prints the **name** of the function being called (use `func.__name__`).

```python
def  name_decorator(func):
    def wrapper():
        print("Function called", func.__name__)
        return func()
    return wrapper
```

```python
@name_decorator
def greet():
    print("Hello!")

greet()
```

```python
@name_decorator
def msg():
    print("New Message")

msg()
```

**Q30.** Combine two decorators: one prints `"Start"`, another prints `"End"` around a function. Check the order of messages.

```python
# TODO: Write your solution for Q30 below
def start_decorator(func):
    def wrapper():
        print("Start")
        return func()
    return wrapper

def end_decorator(func):
    def wrapper():
        result = func()
        print("End")
        return result
    return wrapper
```

```python
@start_decorator
@end_decorator
def greet():
    print("Hello")
```

```python
greet()
```
