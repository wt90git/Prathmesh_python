### Write an adapter function that takes a list of operations and executes add, sub, mul, div based on input. 

### Example: [ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]

```python
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul (a,b):
    print(a*b)
def div(a,b):
    print(a/b)

def ops(**kwargs):
    if kwargs["op"] == "add":
        add(kwargs["a"], kwargs["b"])
    elif kwargs['op'] == 'sub':
        sub(kwargs["a"],kwargs['b'])
    elif kwargs['op'] == 'mul':
        mul(kwargs["a"],kwargs['b'])
    elif kwargs['op'] == 'div':
        div(kwargs["a"],kwargs['b'])
```

```python
lst = [ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]
for i in lst:
    ops(**i)
```

### Write a function process_students(data) that prints total marks and grade for each student.
- Example: [ {"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[60,75,65]} ] 
- Expected Output: Name: A | Total: 240 | Grade: A Name: B | Total: 200 | Grade: B

```python
def student_total(**kwargs):
    s=sum(kwargs["marks"])
    if 0<= s <=300:
        if s>=240:
            grade = "A"
        if s>=200:
            grade = "B"
        if s>=160:
            grade = "C"
        if s>=120:
            grade = "D"
        else:
            grade = "F"
        print("Name:",kwargs["name"],"| Total", s,"| Grade:",grade,"\n")
    else:
        print(kwargs["name"],"-Total marks out of range") 


```

```python
student_info = [ {"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[60,75,65]} ] 
```

```python
for i in student_info:
    student_total(**i)
```

### Write a function highest_salary(emp_list) that returns the employee with the highest salary.

```python
def highest_salary(empl_list):
    highest_salary = 0
    highest_salary_emp = None

    for emp in empl_list:
        if emp['salary'] > highest_salary:
            highest_salary = emp['salary']
            highest_salary_emp = emp['name']

    print("Name:", highest_salary_emp, "| Salary:", highest_salary )
```

### Write out_of_stock(items) that returns all products with stock < 20.

```python
Electronics = [
    {"name": "Laptop", "price": 60000, "stock": 15},
    {"name": "Mouse", "price": 500, "stock": 150},
    {"name": "Keyboard", "price": 1500, "stock": 8},
    {"name": "Monitor", "price": 12000, "stock": 25},
    {"name": "USB Cable", "price": 150, "stock": 50},
    {"name": "Headphones", "price": 2000, "stock": 18}
]
```

```python
def out_of_stock(items):
    out_list = []
    for i in items:
        if i['stock'] < 20:
            out_list.append(i['name'])
    print("Out of stock", out_list)
```
