```python
# 11) Class And Function Design Quetion

orders = [
    {
        "id": 1,
        "student": {"id": "S001", "name": "Aarav"},
        "items": [
            {"name": "Sandwich", "price": 50.0, "qty": 2},
            {"name": "Juice", "price": 30.0, "qty": 1},
        ],
        "payment_method": "Cash"
    },
    {
        "id": 2,
        "student": {"id": "S002", "name": "Neha"},
        "items": [
            {"name": "Pasta", "price": 80.0, "qty": 1},
            {"name": "Juice", "price": 30.0, "qty": 2},
        ],
        "payment_method": "Card"
    },
    {
        "id": 3,
        "student": {"id": "S001", "name": "Aarav"},
        "items": [
            {"name": "Sandwich", "price": 50.0, "qty": 1},
            {"name": "Cookie", "price": 20.0, "qty": 3},
        ],
        "payment_method": "Cash"
    }
]

""" Use Class Based Approach """

# For each order add key "subtotal"
# Subtotal = sum of price * qty for all items.
# print "student_id", "student_name" and "subtotal"

# Add more key "discount_price"
# Discount:
# If subtotal ≥ ₹150, apply 5% discount.
# Else, Original subtotal.
# print "student_id", "student_name" and "discount_price"

# Total tax applied  "tax"
# Tax: Apply 5% tax on the amount after discount.
# example subtotal is 100 then 5 is tax
#  print "student_id", "student_name" and "subtotal", "tax"

# Total revenue for the day (sum of final totals).
# print revenue

```

```python
class Order:
    def __init__(self, data):
        self.data = data
        self.student = data["student"]
        self.items = data["items"]

        self.subtotal = 0
        self.discount_price = 0
        self.tax = 0
        self.final_total = 0
        
    def compute_subtotal(self):
        self.subtotal = sum(item["price"] * item["qty"] for item in self.items)
        self.data["Subtotal"] = self.subtotal

    def calculate_discount(self):
        if self.subtotal >= 150:
            self.discount_price = self.subtotal - (self.subtotal * 0.05)
        else:
            self.discount_price = self.subtotal
        
        self.data["Discounted Price"] = self.discount_price  

    def calculate_tax(self):
        self.tax = self.discount_price * 0.05 
        self.data["Tax"] = self.tax

    def calculate_final_total(self):
        self.final_total = self.discount_price + self.tax
        self.data["Final Total"] = self.final_total  

    def process(self):
        self.compute_subtotal()
        self.calculate_discount()
        self.calculate_tax()
        self.calculate_final_total() 
```

```python
order_objects = []
for order in orders:
    obj = Order(order)
    order_objects.append(obj)

for obj in order_objects:
    obj.process()
```

```python
print("\n---- SUBTOTAL ----")
for order in orders:
    sid = order["student"]["id"]
    name = order["student"]["name"]
    subtotal = round(order["Subtotal"], 2)
    print(sid, name, "=>", subtotal)


print("\n---- DISCOUNT PRICE ----")
for order in orders:
    sid = order["student"]["id"]
    name = order["student"]["name"]
    discount = round(order["Discounted Price"], 2)
    print(sid, name, "=>", discount)


print("\n---- SUBTOTAL + TAX ----")
for order in orders:
    sid = order["student"]["id"]
    name = order["student"]["name"]
    subtotal = round(order["Subtotal"], 2)
    tax = round(order["Tax"], 2)
    print(sid, name, "Subtotal =", subtotal, "Tax =", tax)



revenue = 0
for order in orders:
    revenue += order["Final Total"]

print("\nTOTAL REVENUE:", round(revenue, 2))
```
