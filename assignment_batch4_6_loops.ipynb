## **20. Convert all inner dictionary keys to uppercase.**

````python
employees = {
    'E1': {'name': 'John', 'status': 'active'},
    'E2': {'name': 'Sara', 'status': 'inactive'},
    'E3': {'name': 'Mike', 'status': 'active'},
    'E4': {'name': 'Rita', 'status': 'inactive'}
}
````

````python
for id, info in list(employees.items()):
    for key, value in list(info.items()):
        info.pop(key)
        info[key.upper()] = value
       
    print(id, info)
````

````python
## comprehension
````

````python
up_d = {id: {key.upper(): value for key, value in info.items()} for id, info in employees.items()} 
````

````python
up_d
````
