### 54. Write code to check if a character is a vowel using or.

```python
str = "Adam ate the apple"           
cl_str = str.replace(" ","")
s_s = list(cl_str)
```

```python
print(s_s)
```

```python
for i in s_s:
    if i in ("a,i,e,o,u"):
        print(f"{i} Vowel")
    else:
        print(f"{i} Non Vowel")
```
